```python
import soundfile as sf
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

*Generellt om datainsamlingen:
Alla Annas mätningar innbar att hon skulle anstränga sina muskler tre gånger i rad under inspelningen.
Aron, Johan, Yrsas mätningar var mindre strukturerade, dvs kan vara hårt och snabbt och fler än tre gånger. Dessutom vid tumme och käke spelades dessa in med enklare elektroder som ibland gav rejäla störningar under inspelningen. Detta skedded dock inte på Annas käk och tumme, där var jag extra noga.*

# Biceps

### Arons biceps - snabbt och hårt

```python
data, samplerate = sf.read('1.wav')
```

```python
figure(num=None, figsize=(14, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(data)
plt.ylabel('spike height')
plt.show()
```
